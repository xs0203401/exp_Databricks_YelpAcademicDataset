```python
data_folder = os.path.join('.', 'yelp_data')
json_files = [f for f in os.listdir(data_folder) if f[-4:] == 'json']
json_files
```

```python
df = {}
for f in json_files:
# f = json_files[2]
# if True:
    file_path = os.path.join(data_folder, f)
    print(file_path)
    pd_name = f.split('.')[0].split('_')[-1]
    print(f"--------- Loading into df[{pd_name}] ------------")
    df[pd_name] = pd.read_json(file_path, lines=True, nrows=1_000_000)

print(df.keys())
```
